# 홈페이지 제작을 위한 코드

```python
import pandas as pd

location = './publications_split_submitted.xlsx'
raw = pd.read_excel(location, index_col=None)
raw
```

```python
import copy

df = raw.copy()
df
```

### df['col'] = 'str' + df['col'].astype(str)
를 이용한다

```python
df['etc'] = '<p class="journal">' + df['etc'].astype(str) + '</p>'
df
```

```python
df['author'] = '<p class="name">' + df['author'].astype(str) + '</p>'
df
```

```python
df['name'] = df['name'].astype(str) + '</a></h3>'
df
```

```python
df['url'] = '<h3 class="title"><a href="' + df['url'].astype(str) + '">'
df
```

```python
df['year'] = '<p><strong>Year:</strong> <span class="year">' + df['year'].astype(str) + '</span></p>'
df
```

```python
df['list'] = ''
df
```

```python
df['list'] = '<li>' + df['url'].astype(str) + df['name'].astype(str) + df['etc'].astype(str) + df['author'].astype(str) + df['year'].astype(str) + '</li>'
df
```

```python
df.to_excel('./publications_html.xlsx')
```
